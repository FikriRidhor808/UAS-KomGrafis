```python

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
from google.colab import files
uploaded = files.upload()

# Read the uploaded file
df = pd.read_excel(next(iter(uploaded)))

# Clean the data
df['Purchase Price'] = df['Purchase Price'].replace('[\$,]', '', regex=True).astype(float)
df['Current Price'] = df['Current Price'].replace('[\$,]', '', regex=True).astype(float)
df['Gain/Loss'] = df['Gain/Loss'].replace('[\$,]', '', regex=True).astype(float)
df['Market Value'] = df['Market Value'].replace('[\$,]', '', regex=True).astype(float)
df['Dividend/share'] = df['Dividend/share'].replace('[\$,]', '', regex=True).astype(float)
df['Annual Yield'] = df['Annual Yield'].replace('%', '', regex=True).astype(float)

# a. Distribusi data pada setiap atribut Purchase Price, Current Price, dan Gain/Loss dengan grafik histogram
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.histplot(df['Purchase Price'], kde=True, color='blue')
plt.title('Distribution of Purchase Price')
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
sns.histplot(df['Current Price'], kde=True, color='green')
plt.title('Distribution of Current Price')
plt.xlabel('Current Price')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
sns.histplot(df['Gain/Loss'], kde=True, color='red')
plt.title('Distribution of Gain/Loss')
plt.xlabel('Gain/Loss')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# b. Atribut Stock Name dan Symbol yang memiliki Dividen / Share 5 tertinggi
top_5_dividend = df.nlargest(5, 'Dividend/share')[['Stock Name', 'Symbol', 'Dividend/share']]
plt.figure(figsize=(10, 5))
sns.barplot(data=top_5_dividend, x='Dividend/share', y='Stock Name', palette='viridis')
plt.title('Top 5 Stocks by Dividend per Share')
plt.xlabel('Dividend per Share')
plt.ylabel('Stock Name')
plt.show()

# c. Atribut Stock Name dan Symbol yang memiliki Market Value 5 tertinggi
top_5_market_value = df.nlargest(5, 'Market Value')[['Stock Name', 'Symbol', 'Market Value']]
plt.figure(figsize=(10, 5))
sns.barplot(data=top_5_market_value, x='Market Value', y='Stock Name', palette='magma')
plt.title('Top 5 Stocks by Market Value')
plt.xlabel('Market Value')
plt.ylabel('Stock Name')
plt.show()

# d. Atribut Stock Name dan Symbol yang memiliki Annual Yield 7 tertinggi
top_7_annual_yield = df.nlargest(7, 'Annual Yield')[['Stock Name', 'Symbol', 'Annual Yield']]
plt.figure(figsize=(10, 5))
sns.barplot(data=top_7_annual_yield, x='Annual Yield', y='Stock Name', palette='coolwarm')
plt.title('Top 7 Stocks by Annual Yield')
plt.xlabel('Annual Yield')
plt.ylabel('Stock Name')
plt.show()

top_5_dividend, top_5_market_value, top_7_annual_yield
```
